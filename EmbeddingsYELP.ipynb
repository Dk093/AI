```
wnl=WordNetLemmatizer()
word_list=[]
X=[]
vocab=set()
```

```
for i in range (200):
  X_train.append(df_train.iloc[i][1])
  Y_train.append(df_train.iloc[i][0])
```

```
for i in range(len(X_train)):
  word_list=[]
  review_text = BeautifulSoup(X_train[i]).get_text()
  review_text = re.sub("[^a-zA-Z]"," ", review_text)
  review_text=re.sub(r"[0-9]","digit",review_text)
  review_text=re.sub(r"\'"," ",review_text)
  review_text=re.sub(r"\""," ",review_text)
  review_text=re.sub(r"\\n\\"," ",review_text)
  review_text=re.sub(r"\\"," ",review_text)
  review_text=review_text.lower()
  word_list=review_text.split()
  word_list=[wnl.lemmatize(word) for word in word_list if not word in set(stopwords.words('english'))]
  for word in word_list:
    vocab.add(word)
  word_list=' '.join(word_list)
  X.append(word_list)
```
